```python
import pandas as pd
```

```python
df1 = pd.read_excel('NewstudentList.xlsx')
df1.head()
```

```python
df2 = pd.read_excel('Lil Master copy.xlsx', sheet_name='US-Colleges')
df2.head(22)
```

```python
# Merging
df3 = pd.merge(df1,df2, how ='left', left_on='University Name', right_on = 'University Name' )
df3.head()
```

```python
#exporting data
sheetOne = pd.DataFrame(df3.loc[:,['University Name', 'Category', 'Application_y', 'Early Decision_y','Suppemental/writing']])
file_name = pd.ExcelWriter('NewstudentList.xlsx')

sheetOne.to_excel(file_name, 'US-Colleges', index= False)

file_name.save()

```
